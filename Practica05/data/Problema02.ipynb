```python
import requests
import csv

#crearemos el diccionario paises para luego poder utilizarlo cuando creemos nuestra columna continente.

url = "https://gist.githubusercontent.com/kintero/7d1db891401f56256c79/raw/a61f6d0dda82c3f04d2e6e76c3870552ef6cf0c6/paises.csv"
response = requests.get(url)
if response.status_code == 200:

    content = response.content.decode('utf-8').splitlines()
    
    paises = {}

    csv_reader = csv.reader(content)
    next(csv_reader)
    for row in csv_reader:
        pais = row[1]
        continente = row[6]
      
        paises[pais] = continente
        paises["US"]="América"

    print(paises)
else:
    print("No se pudo obtener el archivo CSV")
```

```python
#Realice al menos 4 renombre de columna.


df_pais = df_pais.rename(columns={
    'country': 'Pais',
    'description': 'Descripcion',
    'points': 'Puntos',
    'price': 'Precio'
})

#3 creación de nuevas columnas según criterio. Deberá crear las columnas que crea conveniente. Ejemplo: Según el país etiquetelos según continente.
df_pais['Continente'] = df_pais['Pais'].map(paises)
df_pais['Descuento'] = df_pais['Precio'] * 0.1 
df_pais['Precio_Final'] = df_pais['Precio'] - df_pais['Descuento'] 



df_pais.head(10)
```

```python
# reporte 01 mejores vinos puntuados por continente
try:
    idx = df_pais.groupby('Continente')['Puntos'].idxmax()
    reporte_mejores_vinos = df_pais.loc[idx, ['Continente', 'title', 'Puntos']]

    reporte_mejores_vinos = reporte_mejores_vinos.sort_values(by='Puntos', ascending=False)
    
except Exception as e:
    print(f"Error en generar el reporte de los continentes con los mejores vinos puntuados: {e}")
reporte_mejores_vinos.head()
```

```python
#reporte 02 promedio de cada pais en puntos y en precio de cada vino
try:
    reporte02 = df_pais.groupby('Pais').agg({
        'Puntos': 'mean',
        'Precio': 'mean'
    }).sort_values(by='Puntos', ascending=False)
except Exception as e:
    print(f"Error en generar el reporte 1: {e}")
reporte02.head(20)
```

```python
#reporte03 bodegas de vinos con los mejores promedios de puntuaciones
try:
     reporte03 = df_pais.groupby('winery').agg({'Puntos': 'mean'}).sort_values(by='Puntos', ascending=False)
       
except Exception as e:
        print(f"Error en generar el reporte 3: {e}")
reporte03.head(10)
```

```python
#reporte04 Promedio de puntos ,precio y cantidad de vinos por continente
try:
    reporte04 = df_pais.groupby('Continente').agg({
        'Puntos': 'mean',
        'Precio': 'mean',
        'title': 'count'
    }).rename(columns={'title': 'Cantidad_de_Vinos'}).sort_values(by='Puntos', ascending=False)

except Exception as e:
    print(f"Error en generar el reporte 4: {e}")
reporte04.head()    
```

```python
try:
    reporte03.to_excel('MejoresPuntuacionesMediasPorBodegas.xlsx', index=True)
except Exception as e:
    print(f"Error al guardar el reporte en Excel: {e}")
```
